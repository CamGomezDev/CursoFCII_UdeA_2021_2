```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t_arr = np.linspace(0., 0.5, 1001)
x_arr = np.linspace(0., 1., 11)

def u(x,t):
    return np.e**(-t*np.pi**2)*np.sin(np.pi*x)

real_small_step = np.zeros((1001, 11))
for i in range(len(real_small_step)):
    for j in range(len(real_small_step[i])):
        real_small_step[i][j] = u(x_arr[j], t_arr[i])
forward_small_step = np.loadtxt("small-step/forward-data.txt")
backward_small_step = np.loadtxt("small-step/backward-data.txt")
crank_nic_small_step = np.loadtxt("small-step/crank-nicolson-data.txt")


fig = plt.figure(figsize=(10,9))
fig.suptitle("Paso pequeño", fontsize=16)

plt.subplot(221)
plt.imshow(real_small_step[-11:], vmin=0., vmax=0.008, extent=[0,1,t_arr[-11],t_arr[-1]], aspect=200)
plt.xlabel('x')
plt.ylabel('t')
plt.title("Real")
plt.colorbar()

plt.subplot(222)
plt.imshow(forward_small_step[-11:], vmin=0., vmax=0.008, extent=[0,1,t_arr[-11],t_arr[-1]], aspect=200)
plt.xlabel('x')
plt.ylabel('t')
plt.title("Forward")

plt.subplot(223)
plt.imshow(backward_small_step[-11:], vmin=0., vmax=0.008, extent=[0,1,t_arr[-11],t_arr[-1]], aspect=200)
plt.xlabel('x')
plt.ylabel('t')
plt.title("Backward")

plt.subplot(224)
plt.imshow(crank_nic_small_step[-11:], vmin=0., vmax=0.008, extent=[0,1,t_arr[-11],t_arr[-1]], aspect=200)
plt.xlabel('x')
plt.ylabel('t')
plt.title("Crank-Nicolson")
```
